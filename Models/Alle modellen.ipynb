```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

eventdf = pd.read_csv("evenementinformatie.csv",sep=';')
berichtdf = pd.read_csv("berichtinteracties_met_sentiment.csv",sep=';')
evenentuserdf = pd.read_csv("evenementaanwezigheid.csv",sep=';')
userdf = pd.read_csv("gebruikersinformatie.csv",sep=';')
```

```python
objs = [berichtdf, userdf]


df = pd.concat(objs, axis=0) #mergen van users en userberichten
```

```python
df
```

```python

compare = [df["GebruikerID"].count(), df["BerichtID"].count()]
labels = ["GebruikerID", "BerichtID"]

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=labels, y=compare)
plt.ylabel("Count")
plt.title("Comparison of GebruikerID and BerichtID Counts")
plt.show()
```
